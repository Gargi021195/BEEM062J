```python
for i in pair:
    URL_A = 'https://min-api.cryptocompare.com/data/histoday?fsym='+i+'&tsym=USD&limit='+lim+ '&api_key='
    URL = URL_A + my_crypto_key
    data = requests.get(URL)
    json_data = data.json()
    table = pd.json_normalize(json_data, 'Data').set_index('time')
    table.index = pd.to_datetime(table.index ,unit='s')
    coins = pd.concat([coins, table.high], axis=1)
coins.columns = pair
```

```python
#Will print the first 50 data of all crytocurriencies
```

```python
coins.columns = pair
coins = coins[pair]
coins
```

```python
coins_lagged=coins.copy()
trailing_window_size = 1
for window in range(1, trailing_window_size +1):
    shifted = coins.shift(window)
    shifted.columns = [x + "_lag" + str(window) for x in coins.columns]
    
    coins_lagged = pd.concat((coins_lagged, shifted), axis=1)
coins_lagged = coins_lagged.dropna()
```

```python
#Using Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df_scaled[['BTC_lag1']]
```

```python
y = df_scaled['BTC']
```

```python
y = y.values.reshape(-1,1)
X = X.values.reshape(-1,1)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
print(lr.intercept_, lr.coef_)
```

```python
#Predicts the value of forecast of BTC
```

```python
forecast_BTC = lr.predict(X)
```

```python
squared_residuals = np.square(forecast_BTC-y)
```

```python
#Mean squared error using linear regression
```

```python
MSE = (np.sum(squared_residuals))/len(y)
MSE
```

From the results of MSE in both the cases, it can be seen that more the number of observations, the less is the value of error.
MSE value of linear regression is lower than MSE value of elasticnet regression, so linear regression is better fit model than elasticnet regression in this case.
